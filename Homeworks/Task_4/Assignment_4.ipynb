## Задача 2

Влияние корреляция между измерениями с ошибкой $\sigma$ на ошибку среднего $(x_1+x_2)/2$?

$$\mathbb{E}[x_1] = \mathbb{E}[x_2] = \mu$$

$$\mathbb{D}[x_1] = \mathbb{D}[x_2] = \sigma ^2$$

$$\rho = \frac{Cov(x_1,x_2)}{\sigma ^2}$$

$$\mathbb{D}[x_1 + x_2] = \mathbb{D}[x_1] + \mathbb{D}[x_2] + 2*Cov(x_1,x_2)$$

$$\mathbb{D}[x_1 + x_2] = 2\sigma^2 + 2\rho  \sigma^2$$

$$D((x_1 + x_2) / 2) = \sigma^2 ( 1 + \rho) / 2$$

$$Ответ:\ \sigma_{\mu} = \sigma \sqrt{\frac{1+\rho}{2}}$$
